```python
group_median = df.groupby('Category')['Review Rating'].transform('median')
df['Review Rating'] = df['Review Rating'].fillna(group_median)
```

```python
df.isnull().sum()
```

```python
df.columns = (
    df.columns
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('purchase_amount_(usd)', 'purchase_amount')
)
```

```python
df.columns
```

```python
age_labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], 4, labels=age_labels)
```

```python
df[['age','age_group']].head(10)
```

```python
freq_map = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].replace(freq_map)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df = df.drop('promo_code_used', axis=1)
```
